```python
a = float(input('Оцените силу вашей душевной боли по шкале от 0 до 10: '))
if a < 0.7:
    print('Кажется, вы испытываете легкое огорчение. Как человек в короткой поездке в метро, забывший дома наушники')
elif 0.7 <= a <= 4.2:
    print('Вы встревожены - почти как человек, которому придется звонить по телефону, и это в эпоху мессенджеров! То самое чувство, когда начинаете нервно репетировать в голове диалог...')
elif 4.2 <= a <= 6.6:
    print('Кажется, тревога с Вами весь день! Такое можно почувствовать, когда Вы готовитесь к экзамену, который не учили. Вам нужно срочно попить чай из ромашки')
elif 6.6 <= a <= 10:
    print ('У Вас бессоница и тремор в руках? Кажется, Ваше состояние близится к критическому. Вам срочно необходима кинотерапия, видео с собачками поднимут Вам настроение')
else:
    print ('Апатия, бессоница и страх. Мы знаем, каково это. Вам срочно нужна психологическая помощь. Пожалуйста, позвоните специалисту прямо сейчас')
```
